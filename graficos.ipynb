```python
# Importando biblioteca
import matplotlib.pyplot as plt

# Dados
instancias = ["10 cidades", "20 cidades", "30 cidades", "40 cidades", "50 cidades"]
custo_otimo = [350, 700, 1050, 1400, 1750]
custo_grasp = [360, 720, 1100, 1470, 1830]
tempo_exec = [0.5, 1.2, 2.5, 4.0, 6.5]

# --- Grfico 1: Comparao de custos ---
plt.figure(figsize=(6,4))
plt.plot(instancias, custo_otimo, marker="o", label="Custo timo (terico)")
plt.plot(instancias, custo_grasp, marker="s", label="Custo GRASP Iterado")
plt.title("Comparao de Custos - TSP com GRASP Iterado")
plt.xlabel("Instncias do Problema")
plt.ylabel("Custo Total do Percurso")
plt.legend()
plt.grid(True)
plt.savefig("grafico1.png")
plt.show()

# --- Grfico 2: Tempo de execuo ---
plt.figure(figsize=(6,4))
plt.bar(instancias, tempo_exec, color="skyblue")
plt.title("Tempo de Execuo - GRASP Iterado no TSP")
plt.xlabel("Instncias do Problema")
plt.ylabel("Tempo (segundos)")
plt.savefig("grafico2.png")
plt.show()
```
